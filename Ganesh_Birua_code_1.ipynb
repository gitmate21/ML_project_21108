#### Spliting features seprated by commas

```python

# Assuming "Experimental Technique" is the column you want to split
column_to_split = "crystallizationMethod"

# Split the values in the specified column by commas and expand into new columns
split_columns = concatenated_data[column_to_split].str.split(',', expand=True)

# Replace empty strings with NaN
split_columns.replace("", np.nan, inplace=True)

# Determine the number of times the column is split
num_splits = split_columns.shape[1]

# Generate new column names based on the number of splits
new_column_names = [f"{column_to_split}_{i+1}" for i in range(num_splits)]

# Rename the split columns
split_columns.columns = new_column_names

# Concatenate the split columns just after the original column
concatenated_data = pd.concat([concatenated_data, split_columns], axis=1)

# Drop the original split column if needed
concatenated_data = concatenated_data.drop(column_to_split, axis=1)

# Display the updated DataFrame
display(concatenated_data)

# Print the shape and size of the updated DataFrame
print("Updated Concatenated data shape is:", concatenated_data.shape)
print("Updated Concatenated data size is:", concatenated_data.size)

concatenated_data.info()
```
